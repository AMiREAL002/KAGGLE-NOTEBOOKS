```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../input/stud-employ/Stud_Employ (2).csv")
df
```

```python
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
predictors = df.iloc[:, 0:4]
target = df.iloc[:,4]
predictors_train, predictors_test,target_train, target_test=train_test_split(predictors, target, test_size=0.3, random_state = 123)
```

```python
predictors_train
```

```python
predictors_test
```

```python
target_train
```

```python
target_test
```

```python
clf=DecisionTreeClassifier()
clf=clf.fit(predictors_train,target_train)
```

```python
clf.predict(predictors_test)
```

```python
from sklearn.tree import plot_tree
ig= plt.figure(figsize=(25,20))
plot_tree(clf, filled=True)
plt.show
```
